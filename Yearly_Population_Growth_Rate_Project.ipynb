```python
import matplotlib.pyplot as plt
import csv
```

```python
# !wget -O Population.csv https://raw.githubusercontent.com/JavohirJalilov/AI-Matplotlib/master/dataset/Yearly%20Population%20Growth%20Rate.csv
```

```python
def read_data(filename):
    """
    Read csv file
    Args:
      file_path(str): file path
    Returns:
      list: data split row
    """
    # WRITE YOUR CODE HERE
    with open(filename) as f:
        data = csv.reader(f)
        return list(data)
```

```python
data_list = read_data("Population.csv")
```

```python
data_list
```

```python
def get_data(data_list):
    """
    Get data from list.

    Args:
      data(list): data split row
    Returns:
      tuple: labels, points
    """
    # WRITE YOUR CODE HERE
    labels = data_list[0]
    years = []
    population = []

    for row in data_list[1:]:
        years.append(int(row[0]))
        population.append(float(row[1]))

    points = [years, population]

    return labels, points
```

```python
labels, points = get_data(data_list)
```

```python
labels
```

```python
def plot_data(labels, points):
    """
    Show data.
    Args:
      gender(list): 0 and 1
      weight(list): weight
      height(list): height
    Returns:
      None
    """

    # WRITE YOUR CODE HERE
    years = points[0]
    population = points[1]
    xlabel = labels[0]
    ylabel = labels[1]

    plt.title(label="Yearly Population Growth Rate", c='blue', size=14, pad=14)
    plt.plot(years, population, c='m')
    plt.xlabel(xlabel=xlabel, c='c', labelpad=14)
    plt.ylabel(ylabel=ylabel, c='c', labelpad=14)
    plt.grid(alpha=.7)
    plt.show()


plt.figure(dpi=100)
plot_data(labels, points)
```
